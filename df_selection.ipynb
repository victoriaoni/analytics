```python
import pandas as pd
```

```python
employees = { 
              'Name': ['John', 'Mark', 'Vicky', 'Daniel', 'Bola', 'Sam', 'Lizzy'],
              'Age': [29, 24, 28, 31, 33, 32, 31],
              'City': ['London', 'Abuja', 'Tokyo', 'Toronto', 'New York', 'Sydney', 'Paris'],
              'Experience': [15, 13, 14, 11, 13, 12, 15]
             }
```

```python
df = pd.DataFrame(employees)
```

```python
df
```

```python
# set_index() to set the index of our df
df = df.set_index('Name')
```

```python
# reset_index() to reset the index
df = df.reset_index()
```

```python
df
```
